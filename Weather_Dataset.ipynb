# **Exploratory Data Analysis(EDA)**

```python
#Explore the weather in count

plt.figure(figsize=(10,5))
sns.set_theme()
sns.countplot(x = 'weather',data = data,palette="rainbow")
plt.title("Count of Weather")
plt.xlabel("weather",fontweight='bold',size=15)
plt.ylabel("Count",fontweight='bold',size=15)
plt.show()
```

```python
#Correlation matrix for numeric dataset
# Select numeric columns
numeric_data = data.select_dtypes(include=np.number)

# Correlation matrix
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

# Time Series analaysis


```python

# Time series analysis of temperature
data['date'] = pd.to_datetime(data['date'])
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['temp_max'], color='red')
plt.title('Time Series of Maximum Temperature')
plt.xlabel('Date')
plt.ylabel('Maximum Temperature')
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.set_theme()
sns.lineplot(x = 'date',y='wind',data=data,color ='purple')
plt.title("Wind")
plt.xlabel("Date",fontweight='bold',size=13)
plt.ylabel("wind",fontweight='bold',size=13)
plt.show()
```

Distribution of Temperature

```python
# Distribution of temperature
plt.figure(figsize=(10, 6))
sns.histplot(data['temp_max'], kde=True, color='lightgreen', edgecolor='red')
plt.title('Distribution of Maximum Temperature')
plt.xlabel('Maximum Temperature')
plt.show()
```

# Pairplot of Wind and weather

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['wind','weather'],axis=1))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.catplot(x='weather',y ='temp_min',data=data,palette = "RdBu")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.catplot(x='weather',y ='temp_max',data=data,palette = "tab10")
plt.show()
```

```python
# Violin Plot on weather Attributes
plt.figure(figsize=(8, 6))
sns.violinplot(data=data[['temp_min','temp_max','wind','precipitation']])
plt.title('Weather Attributes')
plt.xticks()
plt.show()
```
